```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.grid_2d_graph(10,10)
```

```python
nx.draw(g)
```

```python
l = nx.normalized_laplacian_matrix(g).todense()
```

```python
ll = l
plt.figure(figsize=(16,16))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(ll)
    plt.colorbar()
    ll = l @ ll
plt.show()
```

```python
D,P = np.linalg.eigh(l)
plt.imshow(P)
plt.colorbar()
plt.show()
```

```python
plt.imshow(P @ np.diag(D) @ P.T)
plt.colorbar()
plt.show()
plt.imshow(P @ np.diag(D) @ np.linalg.inv(P))
plt.colorbar()
plt.show()
```

```python
plt.imshow(P @ P.T @ np.identity((P.shape[0])))
plt.colorbar()
plt.show()
```
