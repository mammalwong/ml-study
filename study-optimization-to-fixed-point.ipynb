```python
import sympy
import sympy.abc
import matplotlib.pyplot as plt
sympy.init_printing()
```

```python
x = sympy.abc.x
```

```python
F = sympy.Lambda(x, x**3 + 4*x**2 + x + 0)
```

```python
sympy.plot(F(x), ylim=(-10,10))
```

```python
sympy.solve(F(x), x)
```

```python
F(x).diff(x)
```

```python
sympy.plot(F(x).diff(x), ylim=(-10,10))
```

```python
x - (F(x) / F(x).diff(x))
```

```python
sympy.solve(sympy.Eq(x - (F(x) / F(x).diff(x)), x))
```

```python
sympy.plot(x - (F(x) / F(x).diff(x)), x, ylim=(-10,10))
```

```python
sympy.solve(sympy.Eq(F(x), 0), x), \
sympy.solve(sympy.Eq(F(x) + x, x), x)
```

```python
[x_.evalf() for x_ in sympy.solve(F(x), x, seq=True)]
```

```python
[abs((F(x) + x).diff(x).evalf(subs={x: x_})) for x_ in sympy.solve(F(x), x, seq=True)]
```

```python
sympy.plot(F(x) + x, x, ylim=(-10,10))
```

```python
lambda_F = sympy.lambdify(x, F(x) + x)
a = -0.5
for _ in range(100):
    a = lambda_F(a)
a
```

```python
Fint = F(x).integrate(x)
Fint
```

```python
sympy.plot(Fint, xlim=(-1,1), ylim=(-0.02,0.02))
```

### attractor/repellor transform

```python
K = 10
sympy.plot(F(x), ylim=(-10,10), title="f(x) - original function")
sympy.plot(x + F(x)/K, x, ylim=(-10,10), title="x + f(x)/K - tamed attractor transform")
display('hyperbolicities', [abs((x + F(x)/K).diff(x).evalf(subs={x:s})) for s in sympy.solve(F(x))])
sympy.plot(x - F(x)/K, x, ylim=(-10,10), title="x - f(x)/K - tamed repellor transform")
display('hyperbolicities', [abs((x - F(x)/K).diff(x).evalf(subs={x:s})) for s in sympy.solve(F(x))])
```
