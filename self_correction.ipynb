```python
import numpy as np
import matplotlib.pyplot as plt
import keras
import keras.backend as K
```

```python
(X_train, Y_train), (X_test, Y_test) = keras.datasets.mnist.load_data()
Y_train = keras.utils.to_categorical(Y_train, num_classes=10)
Y_test = keras.utils.to_categorical(Y_test, num_classes=10)
```

```python
class model:
    
    def __init__(self):
        self.__dense1 = keras.layers.Dense(4, activation='relu')
        self.__dense1_bn = keras.layers.BatchNormalization()
        self.__dense2 = keras.layers.Dense(4, activation='relu')
        self.__dense2_bn = keras.layers.BatchNormalization()
        self.__dense3 = keras.layers.Dense(10, activation='softmax')
        def sub_model(X_input, previous_Y=None):
            X = X_input
            if previous_Y is not None:
                X_Y = keras.layers.Dense(4)(previous_Y)
                X_Y = keras.layers.LeakyReLU()(X_Y)
                X = keras.layers.Add()([X, X_Y])
            X = self.__dense1(X)
            X = self.__dense1_bn(X)
            X = self.__dense2(X)
            X = self.__dense2_bn(X)
            X = self.__dense3(X)
            return X
        X_input = keras.layers.Input((28,28))
        X_to_sub = keras.layers.Reshape((28,28,1))(X_input)
        X_to_sub = keras.layers.Conv2D(4, (3,3), padding='same', activation='relu')(X_to_sub)
        X_to_sub = keras.layers.MaxPool2D()(X_to_sub)
        X_to_sub = keras.layers.Conv2D(4, (3,3), padding='same', activation='relu')(X_to_sub)
        X_to_sub = keras.layers.MaxPool2D()(X_to_sub)
        X_to_sub = keras.layers.Conv2D(4, (3,3), padding='same', activation='relu')(X_to_sub)
        X_to_sub = keras.layers.MaxPool2D()(X_to_sub)
        X_to_sub = keras.layers.Conv2D(4, (3,3), padding='same', activation='relu')(X_to_sub)
        X_to_sub = keras.layers.MaxPool2D()(X_to_sub)
        X_to_sub = keras.layers.Flatten()(X_to_sub)
        X_to_sub = keras.layers.BatchNormalization()(X_to_sub)
        X_previous_Y = sub_model(X_to_sub)
        for _ in range(3):
            X_previous_Y = sub_model(X_to_sub, X_previous_Y)
        self.__optimizer = keras.optimizers.Adam(clipnorm=5.)
        self.__model = keras.models.Model([X_input], [X_previous_Y])
        self.__model.compile(self.__optimizer, 'categorical_crossentropy', metrics=['accuracy'])
        self.__model.summary()
    
    def fit(self, *args, **kwargs):
        self.__model.fit(*args, **kwargs)
```

```python
m = model()
m.fit(
    X_train, Y_train,
    validation_data=(X_test, Y_test),
    shuffle=True,
    epochs=100)
```
